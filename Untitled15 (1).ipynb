```python
from collections import defaultdict

def read_text_file(path):
    for p in [path, path.with_suffix('.txt'), path.with_suffix('.TXT')]:
        if p.exists():
            try:
                return p.read_text(encoding='latin-1', errors='ignore').strip()
            except:
                pass
    return ""

KEYWORD_MAP = {
    'loch':        'Hole',   'löcher':      'Hole',   'hole':        'Hole',
    'fleck':       'Stain',  'ölfleck':     'Stain',  'öl':          'Stain',
    'schmutz':     'Stain',  'stain':       'Stain',
    'fadenbruch':  'Broken_Thread', 'fadenanzug': 'Broken_Thread',
    'faden':       'Broken_Thread', 'bruch':      'Broken_Thread',
    'thread':      'Broken_Thread', 'broken':     'Broken_Thread',
    'fadenfehler': 'Broken_Thread',
    'webfehler':   'Misweave', 'einzug':    'Misweave',
    'eingewebt':   'Misweave', 'fremdkörper':'Misweave', 'misweave': 'Misweave',
    'kein fehler': 'Background', 'ohne fehler': 'Background',
    'fehlerfreie': 'Background', 'no defect':   'Background',
    'no error':    'Background',
}

def detect_class(text):
    text_lower = text.lower()
    for keyword, cls in KEYWORD_MAP.items():
        if keyword in text_lower:
            return cls
    return None

print("="*60)
print("AUTO-CONFIRMING class mapping from your text files")
print("="*60)

e_class_mapping = {}
sample_path = TILDA_ROOT / "cd2" / "c4" / "r1" / "text"
for e_folder in sorted(sample_path.iterdir()):
    if not e_folder.is_dir():
        continue
    files = sorted(e_folder.iterdir())
    if not files:
        print(f"  {e_folder.name}/  → EMPTY")
        continue
    text     = read_text_file(files[0])
    detected = detect_class(text)
    e_class_mapping[e_folder.name] = detected
    print(f"\n  {e_folder.name}/  →  File: {files[0].name}")
    print(f"  Content : \"{text[:120]}\"")
    print(f"  Detected: {detected}")

print("\n" + "="*60)
print("MAPPING SUMMARY:")
print("="*60)
for e_name, cls in sorted(e_class_mapping.items()):
    print(f"  {e_name}  →  {cls}")
```

```python
import shutil
import cv2
import numpy as np
from pathlib import Path
from collections import defaultdict
from PIL import Image
import random

random.seed(42)

TILDA_ROOT = Path("/content/drive/MyDrive/tilda_CD1+CD2")
BASE       = Path("/content/drive/MyDrive/fabric_defect_yolov8")

E_CLASS_MAP = {
    'e0': None,
    'e1': 'Hole',
    'e2': 'Stain',
    'e3': 'Broken_Thread',
    'e4': 'Misweave',
    'e5': 'SKIP',
    'e6': 'SKIP',
    'e7': 'SKIP',
}

CLASS_ID = {
    'Hole':          0,
    'Stain':         1,
    'Broken_Thread': 2,
    'Misweave':      3,
}
CLASS_NAMES = list(CLASS_ID.keys())

# Clean old classification dataset
cls_dataset = BASE / "dataset_cls"
if cls_dataset.exists():
    shutil.rmtree(cls_dataset)

for split in ['train', 'val', 'test']:
    for cls in CLASS_NAMES:
        (cls_dataset / split / cls).mkdir(parents=True)
print("Classification dataset folders created.")

# Collect all defect images by class
defects = defaultdict(list)
for cd_folder in sorted(TILDA_ROOT.iterdir()):
    if not cd_folder.is_dir(): continue
    for cloth_folder in sorted(cd_folder.iterdir()):
        if not cloth_folder.is_dir(): continue
        for rep_folder in sorted(cloth_folder.iterdir()):
            if not rep_folder.is_dir(): continue
            images_root = rep_folder / "images"
            if not images_root.exists(): continue
            for e_folder in sorted(images_root.iterdir()):
                if not e_folder.is_dir(): continue
                cls = E_CLASS_MAP.get(e_folder.name, 'SKIP')
                if cls not in CLASS_ID:
                    continue
                for ext in ['*.TIF', '*.tif']:
                    defects[cls].extend(e_folder.glob(ext))

print("\nImages found per class:")
for cls, imgs in defects.items():
    print(f"  {cls:<18}: {len(imgs)}")

def load_and_convert(tif_path):
    img  = Image.open(tif_path).convert('RGB')
    arr  = np.array(img)
    lab  = cv2.cvtColor(arr, cv2.COLOR_RGB2LAB)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    lab[:, :, 0] = clahe.apply(lab[:, :, 0])
    arr  = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)
    h, w = arr.shape[:2]
    scale = 640 / max(h, w)
    nw, nh = int(w * scale), int(h * scale)
    resized = cv2.resize(arr, (nw, nh), interpolation=cv2.INTER_LINEAR)
    canvas  = np.full((640, 640, 3), 114, dtype=np.uint8)
    pad_top  = (640 - nh) // 2
    pad_left = (640 - nw) // 2
    canvas[pad_top:pad_top+nh, pad_left:pad_left+nw] = resized
    return canvas

# Stratified 70/15/15 split and write
for cls_name, img_list in defects.items():
    imgs = img_list.copy()
    random.shuffle(imgs)
    n    = len(imgs)
    n_tr = int(0.70 * n)
    n_vl = int(0.15 * n)
    splits = {
        'train': imgs[:n_tr],
        'val'  : imgs[n_tr : n_tr + n_vl],
        'test' : imgs[n_tr + n_vl:]
    }
    for split, files in splits.items():
        out_dir = cls_dataset / split / cls_name
        for tif_path in files:
            canvas   = load_and_convert(tif_path)
            bgr      = cv2.cvtColor(canvas, cv2.COLOR_RGB2BGR)
            out_path = out_dir / f"{tif_path.stem}.jpg"
            cv2.imwrite(str(out_path), bgr, [cv2.IMWRITE_JPEG_QUALITY, 95])

print("\nDataset written:")
for split in ['train', 'val', 'test']:
    for cls in CLASS_NAMES:
        n = len(list((cls_dataset / split / cls).glob("*.jpg")))
        print(f"  {split}/{cls:<18}: {n}")
```

```python
import cv2
import numpy as np
import random
import yaml
from pathlib import Path
from collections import defaultdict
from PIL import Image

random.seed(42)

TILDA_ROOT = Path("/content/drive/MyDrive/tilda_CD1+CD2")
BASE       = Path("/content/drive/MyDrive/fabric_defect_yolov8")

E_CLASS_MAP = {
    'e0': 'Background',     # defect-free → negative examples
    'e1': 'Hole',
    'e2': 'Stain',
    'e3': 'Broken_Thread',
    'e4': 'Misweave',
    'e5': 'SKIP',
    'e6': 'SKIP',
    'e7': 'SKIP',
}

CLASS_ID = {
    'Hole':          0,
    'Stain':         1,
    'Broken_Thread': 2,
    'Misweave':      3,
}

# ── BUILD all_samples (was missing — root cause of NameError) ──────
all_samples = []
for cd_folder in sorted(TILDA_ROOT.iterdir()):
    if not cd_folder.is_dir(): continue
    for cloth_folder in sorted(cd_folder.iterdir()):
        if not cloth_folder.is_dir(): continue
        for rep_folder in sorted(cloth_folder.iterdir()):
            if not rep_folder.is_dir(): continue
            images_root = rep_folder / "images"
            if not images_root.exists(): continue
            for e_folder in sorted(images_root.iterdir()):
                if not e_folder.is_dir(): continue
                cls = E_CLASS_MAP.get(e_folder.name, 'SKIP')
                for ext in ['*.TIF', '*.tif']:
                    for img_path in e_folder.glob(ext):
                        all_samples.append((img_path, cls))

print(f"Total samples collected: {len(all_samples)}")
from collections import Counter
for cls, n in sorted(Counter(c for _, c in all_samples).items()):
    print(f"  {cls:<20}: {n}")

# ── Stratified split ───────────────────────────────────────────────
trainable  = [(p, c) for p, c in all_samples if c in CLASS_ID]
background = [(p, c) for p, c in all_samples if c == 'Background']
skipped    = [(p, c) for p, c in all_samples
              if c not in CLASS_ID and c != 'Background']

print(f"\nTrainable defect samples : {len(trainable)}")
print(f"Background (no defect)   : {len(background)}")
print(f"Skipped (e5/e6/e7)       : {len(skipped)}")

by_class = defaultdict(list)
for item in trainable:
    by_class[item[1]].append(item)
for item in background:
    by_class['Background'].append(item)

train_set, val_set, test_set = [], [], []
for cls, items in by_class.items():
    random.shuffle(items)
    n    = len(items)
    n_tr = int(0.70 * n)
    n_vl = int(0.15 * n)
    train_set += items[:n_tr]
    val_set   += items[n_tr : n_tr + n_vl]
    test_set  += items[n_tr + n_vl:]

print(f"\nSplit → train:{len(train_set)}  val:{len(val_set)}  test:{len(test_set)}")

# ── Conversion helpers ─────────────────────────────────────────────
BBOX_MARGIN = 0.10

def load_and_prep(tif_path):
    img  = Image.open(tif_path).convert('RGB')
    arr  = np.array(img)
    lab  = cv2.cvtColor(arr, cv2.COLOR_RGB2LAB)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    lab[:, :, 0] = clahe.apply(lab[:, :, 0])
    arr  = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)
    h, w = arr.shape[:2]
    scale = 640 / max(h, w)
    nw, nh = int(w * scale), int(h * scale)
    resized = cv2.resize(arr, (nw, nh), interpolation=cv2.INTER_LINEAR)
    canvas  = np.full((640, 640, 3), 114, dtype=np.uint8)
    pad_top  = (640 - nh) // 2
    pad_left = (640 - nw) // 2
    canvas[pad_top:pad_top+nh, pad_left:pad_left+nw] = resized
    return canvas

def make_yolo_label(class_id):
    m  = BBOX_MARGIN
    cx = cy = 0.5
    bw = bh = 1.0 - 2 * m
    return f"{class_id} {cx:.4f} {cy:.4f} {bw:.4f} {bh:.4f}"

def write_split(split_name, items):
    img_dir = BASE / f"dataset/images/{split_name}"
    lbl_dir = BASE / f"dataset/labels/{split_name}"
    img_dir.mkdir(parents=True, exist_ok=True)
    lbl_dir.mkdir(parents=True, exist_ok=True)

    class_counts = defaultdict(int)
    errors = []

    for img_path, cls in items:
        stem    = img_path.stem
        out_jpg = img_dir / f"{stem}.jpg"
        out_lbl = lbl_dir / f"{stem}.txt"
        try:
            canvas = load_and_prep(img_path)
            bgr    = cv2.cvtColor(canvas, cv2.COLOR_RGB2BGR)
            cv2.imwrite(str(out_jpg), bgr, [cv2.IMWRITE_JPEG_QUALITY, 95])
            if cls in CLASS_ID:
                out_lbl.write_text(make_yolo_label(CLASS_ID[cls]))
            else:
                out_lbl.write_text("")   # Background → empty label
            class_counts[cls] += 1
        except Exception as e:
            errors.append((str(img_path), str(e)))

    print(f"\n  {split_name}: {sum(class_counts.values())} images written")
    for cls, n in sorted(class_counts.items()):
        print(f"    {cls:<20}: {n}")
    if errors:
        print(f"  ERRORS ({len(errors)}): {errors[:3]}")

print("Converting and writing dataset...")
write_split('train', train_set)
write_split('val',   val_set)
write_split('test',  test_set)
print("\nDone.")
```
